## 1. Tổng quan dự án

### 1.1 Bối cảnh

Đột quỵ là nguyên nhân gây tử vong đứng thứ hai trên toàn cầu theo WHO. Việc phát hiện sớm các yếu tố nguy cơ có ý nghĩa quan trọng trong phòng ngừa bệnh và giảm tỷ lệ tử vong.

### 1.2 Mục tiêu

Dự án này nhằm:
- **Khám phá dữ liệu** để hiểu đặc điểm của bệnh nhân và các yếu tố nguy cơ đột quỵ
- **Trả lời 6 câu hỏi nghiên cứu** về mối quan hệ giữa các yếu tố với nguy cơ đột quỵ
- **Xây dựng mô hình dự đoán** nguy cơ đột quỵ dựa trên behavioral personas
- **Rút ra insights** có ý nghĩa trong bối cảnh y tế và chăm sóc sức khỏe

### 1.3 Dataset

- **Nguồn:** Kaggle - Stroke Prediction Dataset
- **Kích thước:** 5,110 bệnh nhân × 12 đặc trưng
- **Target:** Stroke (0 = Không đột quỵ, 1 = Có đột quỵ)
- **Class imbalance:** ~95% không đột quỵ, ~5% có đột quỵ
