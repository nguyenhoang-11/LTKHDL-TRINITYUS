1.1 Chủ đề dữ liệu và bối cảnh thực tế
</h2>

Bộ dữ liệu **Stroke Prediction Dataset** tập trung vào bài toán **dự đoán nguy cơ đột quỵ (stroke)** của con người dựa trên các thông tin nhân khẩu học, tiền sử bệnh lý và các chỉ số sức khỏe cơ bản.

Theo Tổ chức Y tế Thế giới (WHO), đột quỵ là **nguyên nhân gây tử vong đứng thứ hai trên toàn cầu** và là nguyên nhân hàng đầu gây tàn tật lâu dài. Việc phát hiện sớm các yếu tố nguy cơ có ý nghĩa quan trọng trong:
- Phòng ngừa bệnh và giảm tỷ lệ tử vong
- Hỗ trợ bác sĩ trong quyết định lâm sàng
- Xây dựng các hệ thống cảnh báo sớm dựa trên dữ liệu

Trong dataset này, **mỗi dòng dữ liệu đại diện cho một bệnh nhân**, phản ánh tình trạng sức khỏe và khả năng xảy ra đột quỵ của cá nhân đó.

1.6 Các đặc trưng trong bộ dữ liệu
</h2>

### Thông tin nhân khẩu học
- `gender`: Giới tính (Male / Female / Other)
- `age`: Tuổi của bệnh nhân
- `ever_married`: Tình trạng hôn nhân (Yes / No)
- `work_type`: Loại hình công việc
- `Residence_type`: Nơi cư trú (Urban / Rural)

### Thông tin sức khỏe
- `hypertension`: Tình trạng tăng huyết áp (0 = Không, 1 = Có)
- `heart_disease`: Tình trạng bệnh tim (0 = Không, 1 = Có)
- `avg_glucose_level`: Mức đường huyết trung bình trong máu (mg/dL)
- `bmi`: Chỉ số khối cơ thể – Body Mass Index (kg/m²)
- `smoking_status`: Tình trạng hút thuốc

### Biến mục tiêu
- `stroke`: Tình trạng đột quỵ  
  - 0: Không bị đột quỵ  
  - 1: Có bị đột quỵ

1.7 Lý do lựa chọn bộ dữ liệu
</h2>

Nhóm lựa chọn bộ dữ liệu này xuất phát từ một thực tế đang ngày càng trở nên rõ ràng
trong xã hội hiện nay: **đột quỵ và các bệnh lý tim mạch không còn là vấn đề chỉ của người cao tuổi**.

Trong nhịp sống hiện đại, đặc biệt là ở **sinh viên và người trẻ**, nhiều thói quen sinh hoạt
được xem là “bình thường” nhưng lại tiềm ẩn rủi ro sức khỏe lâu dài, chẳng hạn như:
- ít vận động do học tập và làm việc kéo dài trước màn hình,
- căng thẳng liên tục bởi áp lực học tập, công việc và tài chính,
- chế độ ăn uống thiếu lành mạnh, ăn nhanh, ăn muộn,
- thiếu quan tâm đến việc kiểm tra sức khỏe định kỳ vì tâm lý còn trẻ và khỏe.

Những yếu tố này thường không gây ra triệu chứng rõ ràng trong ngắn hạn,
khiến nhiều người **chủ quan với các dấu hiệu cảnh báo sớm**.
Chỉ đến khi các vấn đề chuyển hóa như tăng đường huyết, tăng huyết áp
hoặc bệnh tim xuất hiện rõ rệt, nguy cơ đột quỵ mới được nhận ra.

Bộ dữ liệu *Stroke Prediction Dataset* phản ánh tốt bối cảnh này thông qua
việc kết hợp các đặc trưng về tuổi tác, bệnh nền, chỉ số chuyển hóa
và hành vi sinh hoạt.
Dữ liệu cho phép nhóm không chỉ xem xét “ai bị đột quỵ”,
mà còn đặt câu hỏi sâu hơn về **quá trình tích lũy rủi ro theo lối sống**.

Bên cạnh đó, dataset vẫn tồn tại các vấn đề thực tế như thiếu dữ liệu
và mất cân bằng lớp, phản ánh đúng đặc trưng của dữ liệu y tế ngoài đời thực.
Điều này giúp nhóm tiếp cận bài toán không chỉ ở góc độ phân tích kết quả,
mà còn ở góc độ đánh giá chất lượng và giới hạn của dữ liệu.

Vì vậy, bộ dữ liệu này phù hợp với mục tiêu của đồ án,
giúp nhóm xây dựng một câu chuyện dữ liệu gắn liền với đời sống hiện nay,
nhấn mạnh tầm quan trọng của việc **nhận diện sớm nguy cơ sức khỏe
và thay đổi hành vi trước khi các biến cố nghiêm trọng xảy ra**.
