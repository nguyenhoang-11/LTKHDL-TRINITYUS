## 2. Kết quả chính từ EDA

### 2.1 Yếu tố nguy cơ mạnh nhất

Từ phân tích khám phá dữ liệu ([02_EDA](02_EDA.ipynb)), chúng tôi phát hiện:

**1. Tuổi tác - Yếu tố mạnh nhất**
- Nhóm có đột quỵ: tuổi trung bình **67.7 tuổi** (median 71)
- Nhóm không đột quỵ: tuổi trung bình **42.0 tuổi** (median 43)
- **Chênh lệch:** ~25 tuổi

**2. Tăng huyết áp**
- Không tăng huyết áp: 4.0% tỷ lệ đột quỵ
- Có tăng huyết áp: 13.3% tỷ lệ đột quỵ
- **Tăng nguy cơ:** 3.3x

**3. Bệnh tim**
- Không bệnh tim: 4.2% tỷ lệ đột quỵ
- Có bệnh tim: 17.0% tỷ lệ đột quỵ
- **Tăng nguy cơ:** 4.0x

**4. Mức glucose**
- Nhóm có đột quỵ: glucose trung bình **132.5 mg/dL**
- Nhóm không đột quỵ: glucose trung bình **104.8 mg/dL**
- **Chênh lệch:** ~27.7 mg/dL

**5. Tình trạng hôn nhân (proxy cho tuổi)**
- Chưa kết hôn: 1.7% tỷ lệ đột quỵ
- Đã kết hôn: 6.6% tỷ lệ đột quỵ

### 2.2 Các phát hiện khác

- **BMI:** Không có sự khác biệt mạnh giữa hai nhóm (28.8 vs 30.5)
- **Gender:** Không có bằng chứng thống kê mạnh (p > 0.05)
- **Smoking status:** Không tách biệt rõ ràng giữa hai nhóm
- **Correlation:** Không có multicollinearity nghiêm trọng (|r| < 0.7)

### 2.3 Vấn đề dữ liệu

- **Class imbalance nghiêm trọng** (~5% stroke) → Cần SMOTE/class_weight
- **Missing values:** BMI và smoking_status → Cần imputation
- **Outliers:** avg_glucose_level và BMI có nhiều outliers
- **Skewed distributions:** glucose và BMI lệch phải mạnh

### Q1: Young vs Old Stroke Patients - Risk Patterns

**Key Findings:**

**Người trẻ (<40 tuổi):**
- **Yếu tố lối sống đóng vai trò lớn hơn:**
  - Hút thuốc: tỷ lệ cao hơn đáng kể
  - Sống ở đô thị: ~60% vs ~50% (người lớn tuổi)
  - Công việc văn phòng: áp lực cao, ít vận động
- **Bệnh nền:** Tỷ lệ thấp hơn (tăng huyết áp, bệnh tim)
- **Glucose & BMI:** Mức trung bình cao (chuyển hóa kém)

**Người lớn tuổi (>60 tuổi):**
- **Bệnh nền là yếu tố chính:**
  - Tăng huyết áp và bệnh tim phổ biến
  - Tích lũy bệnh lý theo thời gian
- **Yếu tố lối sống:** Vai trò nhỏ hơn

**Kết luận:**
- Người trẻ bị đột quỵ chủ yếu do **lối sống không lành mạnh**
- Người lớn tuổi bị đột quỵ chủ yếu do **bệnh nền tích lũy**
- **Implication:** Cần chiến lược phòng ngừa khác nhau cho từng nhóm tuổi

### Q3: Tương tác Smoking × Age
**Notebook:** [06_Q3_Smoking_Age_Interaction.ipynb](06_Q3_Smoking_Age_Interaction.ipynb)

**Câu hỏi:** Có tương tác giữa hút thuốc và tuổi tác trong việc ảnh hưởng đến nguy cơ đột quỵ?

**Key Findings:**

**Interaction Effect:**
- Ảnh hưởng của hút thuốc **KHÁC NHAU** giữa các nhóm tuổi
- Người trẻ: Hút thuốc có tác động rất mạnh đến nguy cơ đột quỵ
- Người lớn tuổi: Tác động của hút thuốc bị "che lấp" bởi yếu tố tuổi tác

**Stratified Analysis:**
- **Nhóm trẻ (<40):** Hút thuốc tăng nguy cơ đáng kể
- **Nhóm trung niên (40-60):** Tác động hút thuốc vẫn rõ ràng
- **Nhóm cao tuổi (>60):** Tuổi tác là yếu tố chi phối, hút thuốc là yếu tố phụ

**Kết luận:**
- Không thể phân tích smoking status riêng lẻ mà phải kết hợp với age
- Chiến dịch cai thuốc cần tập trung vào **nhóm trẻ và trung niên**
- Ở người cao tuổi, kiểm soát bệnh nền quan trọng hơn

### Q5: Nghịch lý Y tế Đô thị (Urban Penalty)
**Notebook:** [08_Q5_Urban_Rural_Heart_Disease.ipynb](08_Q5_Urban_Rural_Heart_Disease.ipynb)

**Câu hỏi:** Tại sao người đô thị có bệnh tim lại có tỷ lệ đột quỵ cao hơn nông thôn, đặc biệt ở nam giới?

**Key Findings:**

**Urban Penalty - Nam giới bị ảnh hưởng nặng hơn:**
- **Nam đô thị có bệnh tim:** 20.5% stroke rate
- **Nam nông thôn có bệnh tim:** 13.8% stroke rate
- **Chênh lệch:** +6.7% (tăng 48.6%)

**Nữ giới - Không bị urban penalty:**
- **Nữ đô thị có bệnh tim:** 16.9% stroke rate
- **Nữ nông thôn có bệnh tim:** 16.7% stroke rate
- **Chênh lệch:** +0.2% (gần như không khác biệt)

**Tại sao p > 0.05 nhưng vẫn quan trọng?**
- **Cỡ mẫu nhỏ:** Chỉ 47 ca đột quỵ trong nhóm có bệnh tim
- **Effect size lớn:** OR = 1.34 (tăng 34% nguy cơ)
- **Xu hướng rõ ràng:** Nam giới bị ảnh hưởng ~33 lần nữ giới
- **Precautionary Principle:** Trong y tế công cộng, không cần chờ p < 0.05 mới hành động

**Giải thích:**
- Áp lực công việc, stress đô thị ảnh hưởng NAM nhiều hơn
- Ô nhiễm không khí, lối sống ít vận động
- Nam giới có xu hướng ứng phó với stress kém hơn

**Triple Jeopardy Effect:**
- Bệnh tim + Cao tuổi (>60) + Tăng huyết áp
- Đô thị: Nguy cơ cao gấp nhiều lần
- → Hiệu ứng khuếch đại của môi trường đô thị

## 5. Insights và khuyến nghị

### 5.1 Insights chính

**1. Đột quỵ không chỉ là bệnh của người già**
- Người trẻ (<40 tuổi) cũng có nguy cơ nếu lối sống không lành mạnh
- Hút thuốc + Đô thị + Stress công việc → Nguy cơ cao
- **Message:** Phòng ngừa cần bắt đầu từ khi còn trẻ

**2. Lối sống quan trọng hơn chỉ số sinh học**
- BMI/glucose bình thường + Lối sống xấu > BMI/glucose cao + Lối sống tốt
- Chỉ dựa vào xét nghiệm không đủ
- **Message:** Cần đánh giá tổng hợp cả hành vi

**3. Hiệu ứng khuếch đại (Amplification Effect)**
- Nhiều yếu tố nguy cơ cùng lúc → Nguy cơ tăng phi tuyến
- Triple Jeopardy: Bệnh tim + Cao tuổi + Tăng huyết áp → +271%
- **Message:** Cần kiểm soát đồng thời nhiều yếu tố

**4. Môi trường đô thị là yếu tố ẩn**
- Đô thị + Bệnh nền → Nguy cơ cao hơn nông thôn
- Có thể do stress, ô nhiễm, lối sống
- **Message:** Cần can thiệp đặc biệt cho dân số đô thị

### 5.2 Khuyến nghị cho chính sách y tế

**1. Sàng lọc và dự phòng:**
- Mở rộng tiêu chí sàng lọc ngoài BMI, glucose, huyết áp
- Thêm câu hỏi về hút thuốc, stress, hoạt động thể chất
- Chú ý đặc biệt đến nhóm trẻ có lối sống không lành mạnh

**2. Phân tầng nguy cơ:**
- **Nhóm cao nhất:** Đô thị + Bệnh nền + Cao tuổi + Lối sống xấu
- **Nhóm trẻ nguy cơ:** <40 tuổi + Hút thuốc + Văn phòng + Đô thị
- **Nhóm paradox:** Chỉ số sinh học tốt nhưng lối sống xấu

**3. Can thiệp mục tiêu:**
- Người trẻ: Tập trung vào thay đổi hành vi (bỏ thuốc, tăng vận động)
- Người lớn tuổi: Kiểm soát bệnh nền + Duy trì lối sống lành mạnh
- Dân số đô thị: Giảm stress, cải thiện chất lượng môi trường

**4. Truyền thông sức khỏe:**
- Thông điệp: "Chỉ số tốt không đủ, lối sống mới là chìa khóa"
- Nhấn mạnh hiệu ứng khuếch đại của nhiều yếu tố
- Cảnh báo cho người trẻ: Đột quỵ không còn là "bệnh của người già"

### 5.3 Khuyến nghị cho cá nhân

**Nếu bạn là người trẻ (<40 tuổi):**
- ✓ Bỏ thuốc lá ngay lập tức
- ✓ Tăng hoạt động thể chất (ít nhất 30 phút/ngày)
- ✓ Quản lý stress (meditation, yoga, sở thích)
- ✓ Kiểm tra sức khỏe định kỳ (glucose, huyết áp)

**Nếu bạn có bệnh nền (tăng huyết áp, bệnh tim):**
- ✓ Tuân thủ điều trị, uống thuốc đều đặn
- ✓ Kiểm soát chặt chẽ các chỉ số (glucose, BMI, huyết áp)
- ✓ Duy trì lối sống lành mạnh
- ✓ Khám định kỳ 3-6 tháng

**Nếu bạn sống ở đô thị:**
- ✓ Tăng cường hoạt động ngoài trời
- ✓ Giảm stress công việc (work-life balance)
- ✓ Chế độ ăn lành mạnh (tránh fast food)
- ✓ Ngủ đủ giấc (7-8 tiếng/đêm)

## 6. Hạn chế và hướng phát triển

### 6.1 Hạn chế của nghiên cứu

**1. Hạn chế về dữ liệu:**
- **Cross-sectional design:** Không thể xác định quan hệ nhân quả
- **Class imbalance nghiêm trọng:** ~5% stroke → Khó học patterns
- **Missing values:** BMI (4%), smoking_status (30%) → Cần imputation
- **Self-reported data:** Smoking status, work type có thể không chính xác
- **Sample size nhỏ cho nhóm con:** Một số nhóm phân tích có <100 samples

**2. Hạn chế về phương pháp:**
- **Không có temporal information:** Không biết thời điểm đột quỵ, không có follow-up
- **Thiếu biến quan trọng:** Không có thông tin về di truyền, medication, cholesterol, history of TIA
- **Không có validation set độc lập:** Model chưa được test trên external data
- **Chi-square tests với sample nhỏ:** Một số stratified analysis có power thấp

**3. Hạn chế về generalizability:**
- **Nguồn dữ liệu không rõ:** Không biết dataset đến từ quốc gia/bệnh viện nào
- **Không đại diện cho toàn cầu:** Có thể có bias về dân số, địa lý
- **Urban/Rural classification đơn giản:** Không phản ánh đầy đủ mức độ urbanization

### 6.2 Hướng phát triển

**1. Cải thiện dữ liệu:**
- **Longitudinal study:** Thu thập dữ liệu theo thời gian (cohort study)
- **Tăng sample size:** Đặc biệt cho stroke cases
- **Thêm biến quan trọng:**
  - Lab tests: Cholesterol (LDL, HDL), HbA1c, CRP
  - Genetic factors: Family history, genetic markers
  - Lifestyle details: Diet quality, exercise intensity, sleep duration
  - Medical history: Previous TIA, medication, treatment compliance
  - Environmental: Air pollution, noise levels

**2. Cải thiện phương pháp:**
- **Survival analysis:** Cox regression để phân tích time-to-stroke
- **Causal inference:** Propensity score matching, instrumental variables
- **Advanced ML:**
  - Deep learning (Neural Networks) cho pattern recognition
  - Ensemble methods (Stacking) để cải thiện performance
  - Interpretable AI (SHAP, LIME) để explain predictions
- **External validation:** Test model trên dataset từ nguồn khác

**3. Mở rộng phân tích:**
- **Subgroup analysis sâu hơn:**
  - Phân tích theo age-gender-location combinations
  - Identify vulnerable subpopulations
- **Intervention simulation:**
  - What-if analysis: Nếu giảm hút thuốc 20% thì stroke rate giảm bao nhiêu?
  - Cost-effectiveness của các can thiệp
- **Risk calculator development:**
  - Web app để cá nhân tính toán nguy cơ
  - Dashboard cho bác sĩ

**4. Ứng dụng thực tiễn:**
- **Clinical decision support system:**
  - Tích hợp vào EMR (Electronic Medical Records)
  - Real-time alerts cho high-risk patients
- **Mobile health app:**
  - Track lifestyle behaviors
  - Personalized recommendations
  - Gamification để khuyến khích thay đổi hành vi
- **Public health surveillance:**
  - Monitor trends theo thời gian
  - Identify emerging risk patterns
  - Evaluate intervention effectiveness
